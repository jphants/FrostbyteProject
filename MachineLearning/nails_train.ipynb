```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator

base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights="imagenet"
)

base_model.trainable = False

model_nails = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model_eyes = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model_nails.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model_eyes.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
train_gen = ImageDataGenerator(rescale=1./255)

train_data_nails = train_gen.flow_from_directory(
    "dataset/nails/train",
    target_size=(224, 224),
    batch_size=1,
    class_mode="binary"
)

train_data_eyes = train_gen.flow_from_directory(
    "dataset/eyes/train",
    target_size=(224, 224),
    batch_size=1,
    class_mode="binary"
)

model_nails.fit(train_data_nails, epochs=1)
model_eyes.fit(train_data_eyes, epochs=1)
```

```python
model_nails.save("models/anemia_model_nails.keras")
model_eyes.save("models/anemia_model_eyes.keras")

converter = tf.lite.TFLiteConverter.from_keras_model(model_nails)
tflite_model = converter.convert()

with open("models/model_nails.tflite", "wb") as f:
    f.write(tflite_model)

converter = tf.lite.TFLiteConverter.from_keras_model(model_eyes)
tflite_model = converter.convert()

with open("models/model_eyes.tflite", "wb") as f:
    f.write(tflite_model)
```
